```python
# 데이터의 전반적인 내용 확인을 위한 코드 (모든 도시에 대해 진행하고자 하나의 함수를 만들어 진행함.)
def analyze_region(df, region, columns):
    print(f'\n------ {region}에 대한 내용 ------- \n')

    print(f'{region} 의료인력')
    region_df = df[df['dvsd'] == region].sort_values('year')
    display(region_df)

    print(f'\n5년간 {region} 평균 의료인력')
    mean_df = pd.DataFrame(region_df[columns].mean()).transpose()
    mean_df.index = [f'{region}_Average']
    display(mean_df)

    return mean_df

# 사용할 열들
columns = ['dctr','dnst','ordt','nrs','nras','mdth','hmim','chmt','sum']

# 지역 이름들
regions = ['강원', '경기', '경남', '경북', '광주', '대구', '대전', '부산', '서울', '세종', '울산', '인천', '전남', '전북', '제주', '충남', '충북']

for region in regions:
    analyze_region(mp_df, region, columns)
```

```python
# 각 지역별 평균의 행들만 가져와 새로운 데이터 프레임을 만든다.

def analyze_region(df, region, columns):
    region_df = df[df['dvsd'] == region].sort_values('year')
    mean_df = pd.DataFrame(region_df[columns].mean()).transpose()
    mean_df.index = [f'{region}_Average']

    return mean_df  # 평균 데이터프레임을 반환합니다.

# 사용할 열들
columns = ['dctr','dnst','ordt','nrs','nras','mdth','hmim','chmt','sum']

# 지역 이름들
regions = ['강원', '경기', '경남', '경북', '광주', '대구', '대전', '부산', '서울', '세종', '울산', '인천', '전남', '전북', '제주', '충남', '충북']

# 모든 지역별 평균 의료인력 데이터프레임을 모을 빈 데이터프레임을 생성합니다.
all_regions_mean_df = pd.DataFrame()

for region in regions:
    mean_df = analyze_region(mp_df, region, columns)
    all_regions_mean_df = pd.concat([all_regions_mean_df, mean_df])  # 각 지역별 평균 의료인력 데이터프레임을 추가합니다.
```
